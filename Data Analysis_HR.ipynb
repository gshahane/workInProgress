**We can infer following from the statistics:**

**Lets's look at how the data is spread using histogram.**

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure(figsize=(8,4), dpi=300)
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('title')
axes.hist2d(df['satisfaction_level'],df['left'], cmap = df['left'])
```
