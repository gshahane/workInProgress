```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
% matplotlib inline
```

```python
df = pd.read_csv("Revised_Gaurav.csv", encoding = "ISO-8859-1")
df.head()

#df.Country = df.Country.astype('Object')
#df.City/Town = df.City/Town.astype('Object')
df['City VOLY (USD thousands)'] = df['City VOLY (USD thousands)'].astype('float')
df['City DALY lost']= df['City DALY lost'].astype('float')
df['City SVDALY lost (USD millions)'].astype('float')
```

```python
color_labels = np.array(df['Country'].unique())
color_labels
```

```python
df.columns
```

```python
sns.set_context("paper")
sns.set(font_scale=1.5)
plt.figure(figsize=(20, 20))
sns.set_style("whitegrid", {'grid.linestyle': '--'})

fig = sns.lmplot(x = 'City VOLY (USD thousands)', y = 'City DALY lost', hue = 'Country', size = 10, aspect = 1,
                 fit_reg=False, data = df, scatter_kws={"marker": "D","s": 100})

fig = (fig.set_axis_labels("City VOLY in USD Thousands (Capacity)", "City DALY in Years (Risk)").set(ylim=(0, 60000), xlim=(0, 12),
                                                      yticks=[0, 20000, 40000, 60000], xticks=[0, 4, 8, 12]))
plt.title("Capacity Vs Risk")
plt.savefig("CapacityVsRisk.jpg")
```

```python
df = pd.read_csv("data2.csv", encoding = "ISO-8859-1")
df.head()
```

```python
sns.set_context("paper")
sns.set(font_scale=1.5)
plt.figure(figsize=(20, 20))
sns.set_style("whitegrid", {'grid.linestyle': '--'})

fig = sns.lmplot(x = 'City VOLY (USD billions)', y = 'PM 10 Annual mean, ug/m3', hue = 'Country', size = 10, aspect = 1,
                 fit_reg=False, data = df, scatter_kws={"marker": "D","s": 100})

fig = (fig.set_axis_labels("City VOLY in USD Thousands (Capacity)", "PM 10 Annual mean, ug/m3").set(ylim=(0, 600), xlim=(0, 12),
                                                      yticks=[0, 200, 400, 600], xticks=[0, 4, 8, 12]))
plt.title("Capacity Vs Ambient Air Pollution")
plt.savefig("CapacityVsAirPollution.jpg")
```
